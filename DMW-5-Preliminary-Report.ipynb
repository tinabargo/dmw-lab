# Classfying the Players in the NBA Playoffs

## Executive Summary

We classified the NBA players based on their player stats, which was extracted from <a href="https://www.basketball-reference.com/">Basketball Reference</a>. The analysis was limited to players who were able to play in the NBA playoffs from 2010-2019. The goal is to try to surmise the positions of the players in each cluster based on the dominant statistics of the cluster. Agglomerative clustering was used to separate the players into six different clusters. Results show that the benchwarmers are in cluster 1, clusters 2 and 3 are dominated by centers and power forwards, cluster 5 have the best shooters in the game, and cluster 6 have the pointguards. Cluster 4 has no defining characteristic, and were deemed to be all-around players.

## Introduction

Basketball is a team sport wherein two teams (with 5 members each) try to shoot a ball through a defender's hoop while defending the opposing team from shooting in their own hoop. Teams normally assign a position for each player: 
* the point guard (PG) - the one facilitating the team during a play,
* the shooting guard (SG) - usually the best shooter and the best defender,
* the small forward (SF) - normally has an aggressive approach to the basket during ball-handling,
* the power forward (PF) and the center (C) - acting as the team's primary rebounders or shot blockers, or usually receives passes to take inside shots.

The role of a player in the game dictates the statistics he makes on his games. For instance, we can expect the center to rank high in terms of rebounds and blocks, while it is expected for the shooting guard to dominate the field goal efficiency (both 2-pt and 3-pt field goal). However, modern basketball playstyles now employ the concept of "positionless" basketball, which means that roles are somewhat interchangeable. For example, we now see a lot of tall people making 3-pt field goals, and the like.

To test this notion of "positionless" basketball, we looked at the statistics of NBA players. Individual player statistics were extracted from the website <a href="https://www.basketball-reference.com/">Basketball Reference</a>. We scraped the per game stats of all the players that got into the NBA playoffs in the last decade (2010 - 2019). The objective of this project is to classify the NBA players based on their playing statistics, like minutes played, field goal efficiency, steals, blocks, etc. We would like to find out if the classification would naturally separate the roles taken by the players even if the position was dropped from the dataset. Essentially, we want to show if NBA players are indeed playing "positionless" basketball.

## Activity Summary

Data on the raw statistics of each player will be scraped from the website <a href="https://www.basketball-reference.com/">Basketball Reference</a>. The researchers are only interested with the players who participated in the playoffs, expecting a higher level of play in these games compared with the regular season. ### Data Extraction

To start things off, we scraped the URLs of each team that got into the playoffs stage for each year. 

```python
# Initialization:
time = 5
years = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
base_url = 'https://www.basketball-reference.com'

year_links = {}
# Iterate over the years then store the responses in resp list
for yr in years:
    resp = requests.get(f'https://www.basketball-reference.com/playoffs/NBA_{yr}.html', headers=headers)
    sleep(time)
    soup = BeautifulSoup(resp.text, 'lxml')
    tmp_table = soup.findAll('table', {'class':["sortable stats_table", "now_sortable fixed_cols"]})[0]
    rows = [row.findAll('a') for row in tmp_table.findAll('td',{'class':'left'})]
    
    # Get the urls of each playoff team for each year
    links = []
    for i in rows:
        try:
            links.append(base_url+i[0].get('href'))
        except:
            continue
    year_links.update({yr:links})
```

Next, we scraped the bio stats as well as the player URLs. The data gathered for each page was saved in a `CSV` file and stored in a sub-directory in the main directory namely:
* `scraped_data/player_urls/player_urls_{yr}.csv`
* `scraped_data/bio/Bio_Summary-{yr}.csv`

The `bio` statistics consists of physical statistics such as `age`, `height`, `weight` and was taken from the team page of each playoff series.
```python
cnt = 0

# For each playoffs, retrieve the bios of the players and the links to their stats

for yr in years:
    print(f'----Scraping: {yr}')
    
    # Reset the containers after each year
    gen_player_info = []
    players_per_team = []
    
    links = year_links[yr]
    for url in links:
        print(f'Scraping:{url}')
        resp = requests.get(url, headers=headers)
        soup = BeautifulSoup(resp.text, 'lxml')
        tmp_tab = soup.findAll('table', {'id':'roster'})[0]
        gen_player_info.append(pd.read_html(str(tmp_tab)))
        
        # Selects all player anchor tags in the tmp_tab
        for a in tmp_tab.select('a[href*="/players/"]'):
            players_per_team.append((re.findall(fr'teams/(\w+)/{yr}',url)[0], base_url+a.get('href'), a.text))

        sleep(time)
        cnt+=1
        print(cnt)
    
        # Save the links to the player page for a given year
        df_tmp = pd.DataFrame(players_per_team)
        df_tmp.columns = ['team', 'link', 'player']
        df_tmp.to_csv(f'scraped_data/player_urls/player_urls_{yr}.csv')

        # Format then save the bios for a given year.
        list_df = list(map(lambda x: x[0], gen_player_info))
        gen_summary_players = reduce(lambda x, y: x.append(y), list_df)
        gen_summary_players.to_csv(f'scraped_data/bio/Bio_Summary-{yr}.csv')
```

### Cleaning and Formating

In this section the `URL`s that have a `"no_table_found"` tag were filtered out. Rows where the `Season` column was not in years were removed as they are redundant. Rows where the player `"Did Not Play"` were also removed as the researchers are only interested in the raw game statistics. The researchers also added `player name` and `url` columns for better tracking of the game stats.

The `df_fin` dataframe serves a temporary holder for preprocessing, it's contents are eventually saved as a `CSV` file: `scraped_data/player_stats/{player[0]}.csv`; before being overwritten by a different player's `per game` statisitcs.

```python
from functools import reduce

# Remove Players with no per game table.
tmp_holder = list(filter(lambda x: x[1]!= 'no_table_found', per_game_stats))

for title, df in tmp_holder:
    
    # append the player code name from where table came from
    player = re.findall(r'/players/[a-z]/(.*?).html',title)
    df_player = pd.DataFrame(player*len(df[0]), columns=['player'])
    df_fin = df_player.join(df[0])
    
    # append the url where the table came from
    df_link = pd.DataFrame([title]*len(df[0]), columns=['link'])
    df_fin = df_fin.join(df_link)
    
    # choose only year stats
    df_fin['relevance'] = df_fin['Season'].str.contains(r'\d{4}-\d{2}')
    df_fin = df_fin[df_fin['relevance']==True]
    df_fin = df_fin.drop('relevance', axis=1)
    
    # remove rows where the player did not play
    df_fin = df_fin[df_fin['Pos'].str.contains(r'Did Not Play').apply(lambda x: not(x)).tolist()]
    
    # get only total per year stat
        # some players were traded mid year
    try:
        years = df_fin.groupby('Season')['Tm'].count()[
            (df_fin.groupby('Season')['Tm'].count()!=1).to_list()].index.tolist()
        duplicate_seasons = [df_fin[df_fin['Season'] == yr] for yr in years]
        no_dup_seasons = [df_fin[df_fin['Season'] != yr] for yr in years]
        row_list = no_dup_seasons + list(map(lambda x: x[x['Tm']=='TOT'], duplicate_seasons))
        df_fin = reduce(lambda x, y: x.append(y), row_list)
    except:
        pass

    # save the dataframes
    df_fin.to_csv(f'scraped_data/player_stats/{player[0]}.csv', index=False)
```

### Feature Selection

The researchers begin the analysis by doing `feature selection`. Several default columns have high correlation and as such the researchers decided to drop some of the redundant columns. Raw values were preferred over percentages as the percentage does not preserve the frequency of occurence in a given span of time (eg. getting a 100% accuracy in 1 attempt is not comparable to getting a 60% accuracy after 500 attempts).

Here are the features selected for clustering:
* `MP` - minutes played per game
    * players don't get the same amount of play time which will greatly affect the number actions a player can do per game.
* `3P` - 3 point shots that scored
* `3PA` - 3 point attempts made
    * these two together measures the ability of the player to make 3 point shots
    * this denotes the positioning of the player on the court
    * the attempts hints at the player's ability to make / identify opportunities for 3 point shots to happen
* `2P` - 2 point shots that scored
* `2PA` - 2 point attempts made
    * these two together measures the ability of the player to make 2 point shots
    * this also denotes the positioning of the player on the court
    * the attempts hints at the player's ability to make / identify opportunities for 2 point shots to happen
* `FT` - free throws scored
* `FTA` - free throw attempts made
    * these two together measures the ability of the player to make free throw shots
    * the attempts hints at the player's frequency to get fouled
* `ORB`- offensive rebounds made
* `DRB` - defensive rebounds made
    * these two together indicate the player's ability to reach the ball for a rebound
    * this also denotes the positioning of the player on the court
* `AST` - assists made
    * this measures the player's ability to work as a team 
* `STL` - steals made
* `BLK` - blocks made
    * these two statistics show the player's ability to contest the opposing team's offensive play
    * steals hints at ball handling capability
* `TOV` - turnovers made
    * this hints at the ability of the player to gain possession even before the opponet was able to shoot
* `PF` - personal fouls made
    * this hints at a player's playstyle (eg. aggression)

### Clustering

`Agglomerative clustering` was chosen because the researchers want to see how large the difference `delta` is between clusters making it more interpretable. This also serves to show the readers which clusters are similar. The decision on which `method` and `metric` to be used was made by first viewing the dendogram of the different combination of the `method` and `metric`.

```python
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.cluster.hierarchy import fcluster

def multi_dendo(data, met, thresh=0, plevel=5, ignore_ave=True):
    '''Creates multiple dendogram plots.
    
    Parameters:
    -----------
        data: nd array
            Design Matrix to cluster
        met: string
            Distance metric to be used: cityblock, euclidean, or cosine. 
        thresh: Int default 0
            Threshold value to plot as horizontal line
        plevel: Int default 5
            Specify top x levels of the hierarchy to plot 
        ignore_ave: Boolean default True
            Skips the method average if True because of long processing time.
    '''
    fig, ax = plt.subplots(2,3, figsize=(15,10))
    
    print(f'Metric: {met}\n-------')
    try:
        z_ward = linkage(data, method='ward', metric=met, optimal_ordering=True)
        dn_ward = dendrogram(z_ward, ax=ax[0][0], p=plevel, truncate_mode='level')
        ax[0][0].axhline(thresh, color='red')
        ax[0][0].set_title('WARD')
    except:
        print('Ward requires Euclidean metric')
    
    z_single = linkage(data, method='single', metric=met, optimal_ordering=True)
    dn_single = dendrogram(z_single, ax=ax[0][1], p=plevel, truncate_mode='level')
    ax[0][1].axhline(thresh, color='red')
    ax[0][1].set_title('SINGLE')
    
    z_complete = linkage(data, method='complete', metric=met, optimal_ordering=True)
    dn_complete = dendrogram(z_complete, ax=ax[0][2], p=plevel, truncate_mode='level')
    ax[0][2].axhline(thresh, color='red')
    ax[0][2].set_title('COMPLETE')
        
    try:
        z_cent = linkage(data, method='centroid', metric=met, optimal_ordering=True)
        dn_cent = dendrogram(z_cent, ax=ax[1][1], p=plevel, truncate_mode='level')
        ax[1][1].axhline(thresh, color='red')
        ax[1][1].set_title('CENTROID')
    except:
        print('Centroid requires Euclidean metric')
    
    try:
        z_med = linkage(data, method='median', metric=met, optimal_ordering=True)
        dn_med = dendrogram(z_med, ax=ax[1][2], p=plevel, truncate_mode='level')
        ax[1][2].axhline(thresh, color='red')
        ax[1][2].set_title('MEDIAN')
    except:
        print('Median requires Euclidean metric')
        
    if ignore_ave:
        print('Skipping linkage method: Average (takes too long), set ignore_ave = False to override')
    else:
        z_ave = linkage(data, method='average', metric=met, optimal_ordering=True)
        dn_ave = dendrogram(z_ave, ax=ax[1][0], p=plevel, truncate_mode='level')
        ax[1][0].axhline(thresh, color='red')
        ax[1][0].set_title('AVERAGE')
    
    plt.close()
    
    return fig
```

### Agglomerative Clustering: Euclidean Metric

```python
threshold = 8
multi_dendo(X, met='euclidean', thresh=threshold, plevel=5, ignore_ave=False)
```

**Methods ward and complete offer the most balanced clusters for the euclidean metric.**

```python
threshold = 8
multi_dendo(X, met='cityblock', thresh=threshold, plevel=5, ignore_ave=False)
```

**Complete method offers the most balanced clustering for the cityblock metric.**

A delta threshold of `40` was chosen as the researchers saw the large `delta cost` needed to combine the 2 subclusters and deemed it to be too large of a cost. This led to the formation of `6 clusters`.

## Conclusions

Through agglomerative clustering, the researchers were more or less able to identify certain roles assumed by the players during the game.

It appears that the point guards are in `cluster 6`, and they are the ones who are probably holding the ball most of the time. They have the highest number of minutes played, they make the most number of attempts to shoot and are fairly efficient, and they lead the pack in terms of steals, assists and turnovers. They also rank fairly high in terms of personal fouls.

The majority of players in `cluster 2` play the center position. They also rank high in the number of minutes played, they dominate the 2-pt field goal attempts and efficiency, they rank the highest in rebounds and blocks, and they are the second-highest in turnovers. They also rank the highest in terms of personal fouls.

In `cluster 3`, we see a mix of centers and power forwards. They also rank high in terms of 2-pt field goal efficiency, and also do well in rebounds and blocks. They are also likely to be called for personal fouls, since most of the time they play in close contact with other players.

In `cluster 5`, we see the best shooters in the game (based on 2-pt and 3-pt field goal efficiency). They are good defensive players as well, ranking fairly high in terms of steals and assists.

Those in `cluster 1` are most likely the bench warmers, since they rank low in almost all of the statistics considered in this study. They rank pretty high in the field goal efficiency, although it is not clear if this is because of the skill level or due to the fact that they are not given lots of opportunities to shoot the ball in the first place.

Overall it is only players in `cluster 4` that show no clear defining feature stats-wise.

Based on the clustering results, we can somewhat contest the notion of "positionless" basketball, since we are still able to isolate the centers and the pointguards. Although we see a lot of intermixing of these positions in the other clusters, majority of the players still demonstrate the statistics that are expected for the center and the pointguard.
